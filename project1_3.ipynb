```python
companies = df.groupby(['Company', 'Review Date'])['Rating'].mean().unstack()

# خطوط زیر برای استانداردسازی شکل جدول اعمال می‌شوند. برای جواب نهایی آن‌ها تغییر ندهید.
companies.columns.name=None
companies.index.name=None
companies
```

### ranking of companies by their mean of ratings .

```python
mean_ratings = companies.iloc[:,6:14].mean(axis=1)
best_ratings=mean_ratings.reset_index(drop=False).sort_values([0,'index'],ascending=[False,True]).head(10)
best_ratings
```

### Finding best companies and best chocolates.

```python
e=df.Company.isin(best_ratings['index'])
chocolates_to_sell = df[e]
chocolates_to_sell
```

### The sum of Sale 

```python
priceSum = chocolates_to_sell['price(100g)'].sum()
priceSum
```
